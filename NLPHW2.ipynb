```
import nltk
from nltk.corpus import state_union
```

```
state_union.fileids()
```

```
sent=state_union.raw('1945-Truman.txt')
sent
```

```
from nltk.tokenize import word_tokenize
from nltk.collocations import BigramAssocMeasures, BigramCollocationFinder
nltk.download('punkt')
```

```
word_token=word_tokenize(sent)
finder=BigramCollocationFinder.from_words(word_token)
best_bigram=finder.nbest(BigramAssocMeasures().pmi,20)
print('best bigram \n', best_bigram)
```

```
finder.score_ngrams(BigramAssocMeasures().raw_freq)
```

```
from nltk.collocations import TrigramAssocMeasures, TrigramCollocationFinder
trifinder=TrigramCollocationFinder.from_words(word_token)
best_trigram=trifinder.nbest(TrigramAssocMeasures().pmi,20)
print('best bigram \n', best_bigram)
```

```
trifinder.score_ngrams(TrigramAssocMeasures().raw_freq)
```
